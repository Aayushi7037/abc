```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Create student data
students = ["Aman", "Riya", "Rahul", "Sneha", "Karan"]
study_hours = [2, 3, 4, 5, 6]
marks = [50, 60, 65, 75, 85]

# Step 2: Convert data into Pandas DataFrame
df = pd.DataFrame({
    "Student": students,
    "Study Hours": study_hours,
    "Marks": marks
})

print("Student Data")
print(df)

# Step 3: NumPy calculation (average marks)
avg_marks = np.mean(df["Marks"])
print("Average Marks:", avg_marks)

# Step 4: Matplotlib line graph
plt.figure(figsize=(6,4))
plt.plot(df["Study Hours"], df["Marks"], marker='o')
plt.title("Study Hours vs Marks")
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.show()

# Step 5: Seaborn scatter plot
sns.scatterplot(x="Study Hours", y="Marks", data=df)
plt.title("Relationship between Study Hours and Marks")
plt.show()
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
# Step 1: Create student data
students = ["Aayushi", "sharan", "kamal", "divita", "aakriti"]
course = ["data analytics","data science ","data science","web development","cyber security"]
fees = [45,30 , 15, 25, 30]

# Step 2: Convert data into Pandas DataFrame
df = pd.DataFrame({
    "Student": students,
    "course": course,
    "fees": fees
})

print("Student Data")
print(df)

# Step 3: NumPy calculation (average marks)
avg_marks = np.mean(df["Marks"])
print("Average Marks:", avg_marks)

# Step 4: Matplotlib line graph
plt.figure(figsize=(6,4))
plt.plot(df["Study Hours"], df["Marks"], marker='o')
plt.title("Study Hours vs Marks")
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.show()

# Step 5: Seaborn scatter plot
sns.scatterplot(x="Study Hours", y="Marks", data=df)
plt.title("Relationship between Study Hours and Marks")
plt.show()
```
